```python
# drawing a barchat
matplotlib.style.use('seaborn-v0_8-darkgrid')  # and also trying out theme

y = [65, 40, 100]
x = ['APPLE', 'ORANGE','BANANA']

fig, ax = plt.subplots()
# adding labels  & title 

fig.suptitle('Number of fruit Sold')
ax.bar(x, y)
```

```python
## you can also make it horizontal

fig, ax = plt.subplots()

ax.barh(x, y, tick_label=['Apple', 'Orange', 'Banana'], label = 'fruit sold')
ax.legend()

```

```python
# MULTIPLE BAR

y1 = [30, 20, 40]

fig, ax = plt.subplots()

fig.suptitle('Number of fruit Sold')
ax.bar(x, y, label = 'Inventory')
ax.bar(x, y1, label = 'Number sold')

ax.legend(loc = 'best')
```

```python
#let us reduce the width and make the bars side by side
 #from the code above
    # giving it a define size
    
width = 0.5
fig, ax = plt.subplots()
fig.suptitle('Number Of Fruits Sold')
ax.bar(x, y, width = width, label = 'Inventory' )
ax.bar(x, y1, width= width, label = 'Number sold')
ax.legend()
```

```python
# continuation from the above code chucnk
   #manipulating the graph with code
     
x1 = ['APPLE', 'ORANGE','BANANA']    
widthx = 0.25  
fig, ax = plt.subplots()
fig.suptitle('Number Of Fruits Sold')
ax.bar(x1, y1, width= width, label = 'Number sold')
ax.bar(x, y, width = widthx, label = 'Inventory' )
ax.legend()
```

```python
# side by side bars will work with numbers

x2 = [1, 2, 3]
y2 = [15, 20, 15]
y3 = [5, 17.5, 10]

fig, ax = plt.subplots()
fig.suptitle('Jewelry in grams')
labels = ['silver', 'gold', 'diamond']
ax.bar(x2, y2, width=0.4, tick_label = labels, label = 'in stock')

# for y3
x3 = [gram + 0.4 for gram in x2]  # 0.4 was used as spacing because each bar is 0.4wide
ax.bar(x3, y3, width=0.4, tick_label = labels, label = 'sold')

ax.legend()
```

```python
# for the label to be in the center we need one more looping in the x axis

fig, ax = plt.subplots()
fig.suptitle('Jewelry in grams')
labels = ['silver', 'gold', 'diamond']
ax.bar(x2, y2, width=0.4, tick_label = labels, label = 'in stock')

# for y3
x3 = [gram + 0.4 for gram in x2]  # 0.4 was used as spacing because each bar is 0.4wide
ax.bar(x3, y3, width=0.4, tick_label = labels, label = 'sold')

ax.legend()

xticks_labels = [jewelry +0.2 for jewelry in x2]

plt.xticks(xticks_labels, labels)
```

```python
# we can also stack three or more data
  # continuing from the code chunck above

y4 = [7.5, 18, 12.5] 

fig, ax = plt.subplots()
fig.suptitle('Jewelry in grams')
labels = ['silver', 'gold', 'diamond']
ax.bar(x2, y2, width=0.4, tick_label = labels, label = 'in stock')
ax.bar(x2, y4, width=0.4, tick_label = labels, label = 'chunks remainig')
ax.bar(x2, y3, width=0.4, tick_label = labels, label = 'sold')

ax.legend()
```

```python
# let's create dataframe to work with

pie_df = pd.DataFrame([['feeding', 10000],
                     ['transport', 7000],
                     ['power supply', 4000],
                     ['savings' , 6000],
                     ['others', 3000]],
                     columns = ['Expenses', 'Amount ($)'])

pie_df
```

```python
# puting it in pie chart
fig, ax = plt.subplots()

# using the above datafame 
label = pie_df.loc[:, 'Expenses'] # definig the label
pie = pie_df.loc[:, 'Amount ($)'] # specifying the pie value

ax.pie(pie, labels = label, startangle = 60) 
ax.axis('equal') # helps to draw a uniform pie graph
```

```python
fig, ax = plt.subplots()

# from the above code chunck 
label = pie_df.loc[:, 'Expenses'] # definig the label
pie = pie_df.loc[:, 'Amount ($)'] # specifying the pie value

color = ['red', 'yellow', 'blue', 'purple', 'green'] #specifying colors

ax.pie(pie, labels = label, startangle = 60, shadow = True, colors = color)  #adding some formatting
ax.axis('equal')  
```

```python
 fig, ax = plt.subplots()

# from the above code chunck 
label = pie_df.loc[:, 'Expenses'] # definig the label
pie = pie_df.loc[:, 'Amount ($)'] # specifying the pie value

color = ['red', 'yellow', 'blue', 'purple', 'green'] #specifying colors

explode = [0.05, 0.05, 0.05 , 0.05, 0.05] #spacing the pie
ax.pie(pie, labels = label, startangle = 60, shadow = True, colors = color, explode=explode)  #adding some formatting
ax.axis('equal')  
```

```python
fig, ax = plt.subplots()
fig.suptitle('Monthly Expenses on Salary ($30,000)')

# from the above code chunck 
label = pie_df.loc[:, 'Expenses'] # definig the label
pie = pie_df.loc[:, 'Amount ($)'] # specifying the pie value

color = ['red', 'yellow', 'blue', 'purple', 'green'] #specifying colors

explode = [0, 0, 0, 0, 0.1]  # spacing only the smallest pie

ax.pie(pie, labels = label, startangle = 60, shadow = True, colors = color, 
       explode=explode, autopct = '%1.0f%%')  #adding percentage to it
ax.axis('equal')  
```
