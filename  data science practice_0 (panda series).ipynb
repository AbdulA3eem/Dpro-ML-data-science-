```python
# we can concatenate and join more dataframe together
```

```python
# lets us manually randomly create two dataframe for this:
D1 = pnd.date_range(start = '2024/05/12', end = '2024/05/18' )
df1 = pnd.DataFrame({"Sales $": [64, 52, 77, 89, 39, 60, 93]}, 
                  index = D1)

df1
```

```python
# let's us create another one :

df2 = pnd.DataFrame({"purchases $": [60, 63, 69, 80, 40, 55, 93]},
                   index = D1)

df2
```

```python
# now using concat function to join them together

df12 = pnd.concat([df2, df1], axis = 1)
df12
```

```python
# let get some caculations from it 

df12.loc[:, "profit"] = df12.loc[:, "Sales $"] - df12.loc[:, "purchases $"]

df12
```

```python
# you can get some stat with it eg (correlation)
```

```python
df12.corr()
```
