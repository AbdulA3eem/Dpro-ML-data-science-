```python
# [Problem 1] Creating random numbers
```

```python
import numpy as np

# Setting the seed for fixed value
np.random.seed(0)

# Mean and covariance matrix
mean1 = [-3, 0]
covariance = [[1.0, 0.8], [0.8, 1.0]]

# Generate 500 samples
data1 = np.random.multivariate_normal(mean1, covariance, 500)

data1
```

```python
# [Problem 2] Visualization by scatter plot
```

```python
import matplotlib.pyplot as plt

# creating a Scatter plot
fig, ax = plt.subplots()
fig.suptitle('Scatter Plot of Data1')
ax.scatter(data1[:, 0], data1[:, 1])

plt.xlabel('X1')
plt.ylabel('x2')
plt.show()
```

```python
# [Problem 3] Visualization by histogram
```

```python
# Create histograms for each dimension
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Histogram for X1 dimension
ax[0].hist(data1[:, 0], bins=30, color='blue')
ax[0].set_title('Histogram of X1')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('Frequency')

# Histogram for X2 dimension
ax[1].hist(data1[:, 1], bins=30, color='green')
ax[1].set_title('Histogram of X2')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Frequency')

# Align the range of the horizontal axis for both histograms
x_min = min(data1[:, 0].min(), data1[:, 1].min())
x_max = max(data1[:, 0].max(), data1[:, 1].max())
ax[0].set_xlim(x_min, x_max)
ax[1].set_xlim(x_min, x_max)

plt.tight_layout()
plt.show()
```

```python
# [Problem 4] Addition of data
```

```python
# Mean for the new data
mean2 = [0, -3]

# Generate 500 new samples
data2 = np.random.multivariate_normal(mean2, covariance, 500)

# Combined scatter plot
fig, ax=plt.subplots()
fig.suptitle('Scatter Plot of Data1 and Data2')

ax.scatter(data1[:, 0], data1[:, 1], label='Data1 ')
ax.scatter(data2[:, 0], data2[:, 1], label='Data2 ')
 
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
# [Problem 5] Data combination
```

```python
# Combining the datasets
combined_data = np.vstack((data1, data2))

combined_data.shape
```

```python
combined_data
```

```python
# [Problem 6] Labeling
```

```python
# Creating labels for the dummy data

labels1 = np.zeros(500)  # Label for data1
labels2 = np.ones(500)   # Label for data2

# Combining both labels
combined_labels = np.hstack((labels1, labels2))

# Combining the data and labels
final_data = np.hstack((combined_data, combined_labels.reshape(-1, 1)))

final_data
```
