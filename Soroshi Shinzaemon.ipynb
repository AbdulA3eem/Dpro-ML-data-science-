```python
# PROBLEM 2
def calculate_rice_grains(n):
    grains = 1  # Starting with 1 grain of rice on the first day
    rice_grains_list = [grains]
    for _ in range(1, n):
        grains *= 2
        rice_grains_list.extend([grains, grains])
    return rice_grains_list

def rice_grains_story(days):
    rice_grains_list = calculate_rice_grains(days)
    return rice_grains_list

# lets assume i have take this course for 18 days
story_days = 18
story_result = rice_grains_story(story_days)
print('Rice grains story for', story_days, 'days:')
print(story_result)

# list for problem 2

grains = 1  # 1 on the first day 


rice_grains_growth = []

rice_grains_growth.append(grains)

days = story_days

for _ in range(days):
    total_grains = rice_grains_growth[-1] * 2
    rice_grains_growth.append(total_grains)

len(rice_grains_growth)

```

```python
# problem 2 visualization

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 5))
plt.title("Rice Grains Growth")
plt.xlabel("number of days")
plt.ylabel("number of rice")
plt.plot(rice_grains_growth) 
plt.grid(True)

plt.show()
```

```python
# problem 3

def calculate_days_of_sustenance(rice_grains, num_people, grains_per_day):
    total_grains_needed = num_people * grains_per_day
    days_of_sustenance = rice_grains / total_grains_needed
    return days_of_sustenance


# Given values 
num_people_in_class = 60  # number of people in class was guessed from the number of the whatsapp participant
course_duration = 30  # days # course duration estimated to be 1 month
rice_grains_received = calculate_total_grains(course_duration)  # Calculate the total grains over the course duration
grains_per_person_per_day = 10000 # according to surbivalfreedom and medical news today


# Calculate how many days the class can live with the provided rice
days_of_sustenance = calculate_days_of_sustenance(rice_grains_received, num_people_in_class, grains_per_person_per_day)

print(f"The class of {num_people_in_class} people can live for approximately {days_of_sustenance:.2f} days with the provided rice.")
```
